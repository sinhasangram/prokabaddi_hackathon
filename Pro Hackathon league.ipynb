```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
final_play[['Matches Played','Win','Lost','Tie']]=scaler.fit_transform(final_play[['Matches Played','Win','Lost','Tie']])
```

```python
y_train=final_play.pop('Win %')
X_train=final_play
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
```

```python
# step-1: create a cross-validation scheme
folds = KFold(n_splits=5,shuffle=True,random_state=100)

# step-2: specify range of hyperparameters to tune
hyper_params=[{'n_features_to_select':list(range(1,21))}]

# step-3: perform grid search
# 3.1 specify model
lm=LinearRegression()
lm.fit(X_train,y_train)
rfe=RFE(lm)

# 3.2 call GridSearchCV()
model_cv=GridSearchCV(estimator=rfe,param_grid=hyper_params,scoring='r2',
                     cv=folds,verbose=1,return_train_score=True,refit="precision_score")

# fit the model
model_cv.fit(X_train,y_train)
```

```python
cv_results=pd.DataFrame(model_cv.cv_results_)
cv_results
```

```python
# plotting cv results
plt.figure(figsize=(16,6))

plt.plot(cv_results['param_n_features_to_select'],cv_results['mean_train_score'])
plt.plot(cv_results['param_n_features_to_select'],cv_results['mean_test_score'])
plt.xlabel('Number of features')
plt.ylabel('r-squared')
plt.title('Optimal Number of Features')
plt.legend(['test score', 'train score'], loc='upper left')
plt.show()
```

```python
# final model
n_features_optimal = 14

lm1=LinearRegression()
lm1.fit(X_train,y_train)
rfe=RFE(lm1,n_features_to_select=n_features_optimal)
model=rfe.fit(X_train,y_train)
```
